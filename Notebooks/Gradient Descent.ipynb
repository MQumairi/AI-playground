### Gradient Descent Algorithm 

```python
def gradient_descent (derivative_func, initial_guess, multiplier= 0.02, 
                      precision= 0.001, max_iter= 300):
    
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]
    
    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        #update step
        new_x = previous_x - multiplier * gradient
        
        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))
        
        if step_size < precision:
            break
            
    return new_x, x_list, slope_list
```

### Example 1 - A simple cost function

#### $$f(x) = x^2 + x + 1$$

```python
def f(x):
    return x**2 + x + 1
```

```python
#Make data
x_1 = np.linspace(start=-3, stop=3, num=500)
```

```python
#Plot
plt.plot(x_1, f(x_1))
plt.xlim([-3, 3])
plt.ylim([0, 10])
plt.xlabel("X")
plt.ylabel("f(x)")
plt.show()
```

```python
def df(x): 
    return 2*x + 1
```

```python
#Plot

#Set figure size
plt.figure(figsize=[15, 5])

#First plot
plt.subplot(1, 2, 1)

plt.plot(x_1, f(x_1))
plt.xlim([-3, 3])
plt.ylim([0, 10])
plt.xlabel("X")
plt.ylabel("f(x)")

#Second plt
plt.subplot(1, 2, 2)
plt.plot(x_1, df(x_1))
plt.grid()

plt.show()

```

```python
# Calling Gradient descent
local_min, list_x, deriv_list = gradient_descent(derivative_func=df, 
                                                 initial_guess=-3,
                                                 multiplier=0.1)

#Plot

#Set figure size
plt.figure(figsize=[15, 5])

#First plot
plt.subplot(1, 2, 1)

plt.plot(x_1, f(x_1))
plt.xlim([-3, 3])
plt.ylim([0, 10])
plt.xlabel("X")
plt.ylabel("f(x)")

plt.scatter(list_x, f(np.array(list_x)), color='red')

#Second plt
plt.subplot(1, 2, 2)
plt.plot(x_1, df(x_1))
plt.grid()

plt.show()
print("Local minimum is at " + str(local_min))
```

## Example 2: Multiple Minima
$$g(x) = x^4 -4x^2+5$$
$$dg(x) = 4x^3 - 8x$$

```python
x_2 = np.linspace(-2, 2, 1000)
```

```python
#g(x)

def g(x):
    return pow(x, 4) - 4 * pow(x, 2) + 5
```

```python
#dg(x)

def dg(x):
    return (4 * pow(x, 3)) - (8 * x)
```

```python
plt.figure(figsize=[15,5])

#first plot
plt.subplot(1, 2, 1)
plt.xlim(-2, 2)
plt.ylim(-0.5, 5.5)

plt.plot(x_2, g(x_2))

#second plot
plt.subplot(1, 2, 2)
plt.xlim(-2, 2)
plt.ylim(-6, 8)
plt.grid()
plt.plot(x_2, dg(x_2))
```
